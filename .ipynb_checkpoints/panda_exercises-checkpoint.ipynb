## Exercise 2
Sorting/Filtering dataframe

```python
product_data = {
    'Product': ['Laptop', 'Headphones', 'Smartphone', 'Washing Machine', 'Fridge'],
    'Price': [300, 50, 800, 1200, 600],
    'Rating': [4.7, 3.8, 4.5, 3.2, 4.9]
}

products = pd.DataFrame(product_data)
products
```

```python
high_rated = products[products["Rating"] > 4.0]
high_rated
```

```python
affordable = products[products["Price"] <= 50]
affordable
```

```python
expensive_high_rated = products[(products["Price"] > 100) & (products["Rating"] > 4.5)]
expensive_high_rated
```

```python
filtered_sorted_products = products[(products["Price"] > 30) & (products["Rating"] > 3.5)]
filtered_sorted_products = filtered_sorted_products.sort_values(by="Rating", ascending=False)
# In one line
# filtered_sorted_products = products[(products["Price"] > 30) & (products["Rating"] > 3.5)].sort_values(by="Rating", ascending=False)
filtered_sorted_products
```

## Exercise 3

```python
student_data = {
    'StudentID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'MathGrade': [95, 65, 45, 75, 60, 58, 90, 88],
    'EnglishGrade': [85, 72, 85, 60, 65, 78, 85, 70],
    'ScienceGrade': [92, 60, 58, 92, 80, 70, 90, 83]
}

student_grades = pd.DataFrame(student_data)
student_grades
```

```python
struggling = student_grades[
    (student_grades[['MathGrade', 'EnglishGrade', 'ScienceGrade']] <= 60).any(axis=1)
]

# Same thing but with range instead of specifying column names
struggling = student_grades[
    (student_grades.iloc[:, 2:] <= 60).any(axis=1)
]
struggling
```

```python
top_performers = student_grades[
    student_grades[['MathGrade', 'EnglishGrade', 'ScienceGrade']].mean(axis=1) > 85
]
# Same thing but with range instead of specifying column names
top_performers = student_grades[
    student_grades.iloc[:, 2:].mean(axis=1) > 85
]

top_performers
```

```python
# Honor roll eligible students (Math and Science grades both 90 and above)
honor_roll_eligibility = [
    (student_grades['MathGrade'] >= 90) & (student_grades['ScienceGrade'] >= 90)
]
honor_roll_eligibility
```

```python
honor_roll_eligible = student_grades[
    (student_grades['MathGrade'] >= 90) & (student_grades['ScienceGrade'] >= 90)
]
honor_roll_eligible
```

```python
top_math = student_grades.nlargest(1, 'MathGrade')
top_math
```

```python
top_english = student_grades[
    student_grades['EnglishGrade'] == student_grades['EnglishGrade'].max()
]
top_english
```

```python
max_science_grade = student_grades['ScienceGrade'].max()
top_science = student_grades[student_grades['ScienceGrade'] == max_science_grade]
top_science
```

## Exercise 1

```python
new_student_data = {
    'StudentID': [9, 10, 11, 12, 13],
    'Name': ['Ian', 'Jenny', 'Kurt', 'Lila', 'Mike'],
    'MathGrade': [71, 49, 85, 45, 48],
    'EnglishGrade': [74, 57, 86, 53, 65],
    'ScienceGrade': [91, 96, 93, 100, 55]
}

new_students = pd.DataFrame(new_student_data)
student_grades = pd.concat([student_grades, new_students], ignore_index=True)
student_grades.head(5)
```

```python
student_grades.iloc[:, 2:4].plot(kind="scatter", x="MathGrade", y="EnglishGrade", title="Math and English Grade Correlation")
# Second way of creating the same plot specifying the column names
# student_grades[["MathGrade", "EnglishGrade"]].plot(kind="scatter", x="MathGrade", y="EnglishGrade", title="Math and English Grade Correlation")
#plt.Show()
```

```python
student_grades["ScienceGrade"].plot(kind="bar", x="StudentID", y="Grade", title="Science Grade Distribution")
```

```python
# Adding the "Gender" column
gender_mapping = {
    'Alice': 'Female', 'Bob': 'Male', 'Charlie': 'Male', 'David': 'Male',
    'Eva': 'Female', 'Frank': 'Male', 'Grace': 'Female', 'Helen': 'Female',
    'Ian': 'Male', 'Jenny': 'Female', 'Kurt': 'Male', 'Lila': 'Female',
    'Mike': 'Male'
}
student_grades['Gender'] = student_grades['Name'].map(gender_mapping)
student_grades
```

```python
# Rearranging columns
# 1st way of moving column
cols = list(student_grades.columns)
rearranged_student_grades = student_grades[cols[0:2] + [cols[-1]] + cols[2:5]] # cols[-1] is the first from end

#2nd way of moving column
gender_col = student_grades.pop('Gender') # Remove 'Gender' column
student_grades.insert(2, 'Gender', gender_col) # Insert 'Gender' column back 
student_grades
```

```python
rearranged_student_grades
```

```python
student_grades["Gender"].value_counts().plot(kind="pie", title="Student Gender Distribution", autopct='%1.1f%%', ylabel="", legend=True)
plt.show()
```

```python
# max_grade = student_grades['MathGrade'].max()
# bins = [0] + list(range(50, max_grade + 20, 20))
bins = [0, 50, 70, 90, 100]
labels = ['0-50', '50-70', '70-90', '90-100']
pd.cut(student_grades["MathGrade"], bins=bins, labels=labels).value_counts().sort_index().plot(kind="bar")
plt.title("Distribution of Math Grades in Ranges")
plt.xticks(rotation=0)
plt.show()
```

## Exercise 2

```python
employee_data = {
    "EmployeeID": ["1", "2", "3", "4", "5", "6", "7", "8", "9", "0"],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward", "Fiona", "George", "Hannah", "Ian", "Jane"],
    "Salary": [60000, 75000, 50000, 82000, 95000, 72000, 58000, 67000, 89000, 61000],
    "YearsOfExperience": [5, 8, 3, 10, 12, 7, 4, 6, 11, 5],
    "PerformanceScore": [4, 3, 5, 2, 4, 3, 5, 4, 3, 4],
}

# Create the DataFrame
employee_df = pd.DataFrame(employee_data)
employee_df
```

```python
employee_df.sort_values("YearsOfExperience").plot(kind="line", x="YearsOfExperience", y="Salary", marker="o", title="Salary Trend", figsize=(6, 4))
plt.xlabel("Years of Experience")
plt.ylabel("Salary (EUR)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5)) 
sorted_employees = employee_df.sort_values("YearsOfExperience")
plt.plot(sorted_employees["YearsOfExperience"], sorted_employees["Salary"], marker="o")
plt.title("Salary Trend")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (EUR)")
plt.grid(True)
plt.show()
```

```python
employee_df["PerformanceScore"].value_counts().plot(kind="pie", title="Distribution of Performance Scores", autopct='%1.1f%%', ylabel="", legend=True)
```

```python
bins = [0, 50000, 75000, 100000]
labels = ['0-50000', '50000-75000', '75000-100000']
pd.cut(employee_df["Salary"], bins=bins, labels=labels).value_counts().sort_index().plot(kind="bar")
plt.title("Salary Range Distribution")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

## Exercise 3

```python
SalesData = {
    "ProductID": [101, 102, 103, 104, 105, 106],
    "ProductName": ["Laptop", "Smartphone", "Tablet", "Headphones", "Smartwatch", "Monitor"],
    "QuantitySold": [50, 120, 80, 150, 60, 30],
    "Revenue": [50000, 60000, 32000, 7500, 12000, 9000],
    "Region": ["North", "South", "East", "West", "North", "East"]
}

sales_df = pd.DataFrame(SalesData)
sales_df
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(
    sales_df["QuantitySold"],
    sales_df["Revenue"],
    s=sales_df["Revenue"] / 100,  # Scale the bubble size based on Revenue
    zorder=3 # Ensure bubbles are on top
)
plt.grid(True, alpha=0.7)
```

```python
plt.bar(sales_df["Region"], sales_df["Revenue"], width=0.5, zorder=3, color="limegreen")
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.grid(True, axis='y', alpha=0.7)
```

```python
product_sales = sales_df.groupby("ProductName")["QuantitySold"].sum()

ax = product_sales.plot(kind="pie", autopct='%1.1f%%', legend=True, colors=plt.cm.Set2.colors, figsize=(6,6))

ax.legend(loc='center left', bbox_to_anchor=(0.92, 0.5))
plt.title("Sales by Product")
plt.ylabel("")
plt.show()
```

```python
product_sales = sales_df.groupby("ProductName")["QuantitySold"].sum()

plt.pie(
    product_sales,   
    labels=product_sales.index,   # Use product names as labels
    autopct='%1.1f%%',            # Format the percentage labels
    startangle=90,                # Start the pie chart at 90 degrees
    colors=plt.cm.Set3.colors,    # Color palette for the pie chart
    wedgeprops={'edgecolor': 'black'}  # Add a black edge to the slices for clarity
)
plt.title("Proportion of Sales by Product (Quantity Sold)")
plt.legend(product_sales.index, loc='lower left', bbox_to_anchor=(0.92, 0.63))

plt.show()
```
